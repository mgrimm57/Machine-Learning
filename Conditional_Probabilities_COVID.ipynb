Let us define a random variable $T$= "COVID test results" with values $t=0,1$, and a random variable $G$=COVID, with values $g=0,1$. There are several available tests for COVID, including "antigen" tests and "RT-PCR" tests. We will use the "antigen" test.

I will compile the confusion matrix for this test, namely the rates for: hit=$P(t=1|g=1)$, miss=$P(t=0|g=1)$, false positive=$P(t=1|g=0)$, correct rejection=$P(t=0|g=0)$.

Let us assume that a good estimate for the occurrence of COVID in the local Lane County population $P(g=1)$ is given by the fraction of people infected per 100k, as reported at the [Johns Hopkins Resource Center](https://coronavirus.jhu.edu/us-map).

Q: What is the probability of having COVID if you tested positive $P(g=1|t=1)$?


### Answer:

#### Deriving our probability
First, we are going to find an equation for our desired conditional probability $P(g=1|t=1)$. So using bayes theorem...

$$ P(g=1|t=1) = \frac{P(t=1|g=1)*P(g=1)}{P(t=1)}$$

Since we can not find a value for $ P(t=1) $, we will use the marginal probability which leads us to...

$$ P(g=1|t=1) = \frac{P(t=1|g=1)*P(g=1)}{\sum_{g'=0,1} P(t=1|g=g')*P(g=g')} $$

This simplifies to

$$ P(g=1|t=1) = \frac{P(t=1|g=1)*P(g=1)}{P(t=1|g=0)*P(g=0)+P(t=1|g=1)*P(g=1)} $$


Now, lets establish our probabilities and where they came from. From the John Hopkins Resource Center, I found that 6,833 people were infected per 100k. This gives us...

$$ P(g=1) = 0.0683 $$

Therefore,

$$ P(g=0) = 1-P(g=1) = 0.9317 $$

From Cochrane [link](https://www.cochrane.org/CD013705/INFECTN_how-accurate-are-rapid-tests-diagnosing-covid-19), we find some important conditonal probabilities. From this website, they breakdown their results into two categories (people with COVID symptoms and people with no COVID symptoms). I chose to take a look at both probabilites.

---
#### People who have COVID symptoms:
From the website, we found...
$$ P(t=1|g=1)= 0.72 $$
$$ P(t=0|g=0)= 0.995 $$

Therefore, we will assume that...

$$P(t=0|g=1)=0.28$$
$$P(t=1|g=0)=0.005$$

Now, we can complete the confusion matrix.

| Hit: 0.72 |  Miss: 0.28 |
|--- |---|
| **False Postive: 0.005** |**Correct Recognition: 0.995** |

Now we can use our probabilites and the expression above to find our desired probability

$ P(g=1|t=1)_{symptom} = \frac{(0.72)*(0.0683)}{(0.005)*(0.9317)+(0.72)*(0.0683)} $


$ P(g=1|t=1)_{symptom} = 0.9135 $

**Here we found that the probability of having COVID if you tested positive and are experiencing symptoms is roughly 91%**

---
#### People who did not have COVID symptoms:
From the website, we found...
$$ P(t=1|g=1)= 0.58 $$
$$ P(t=0|g=0)= 0.989 $$

Therefore, we will assume that...

$$P(t=0|g=1)=0.42$$
$$P(t=1|g=0)=0.011$$

Now, we can complete the confusion matrix.

| Hit: 0.58|  Miss: 0.42 |
|--- |---|
| **False Postive: 0.011** |**Correct Recognition: 0.989** |

Now we can use our probabilites and the expression above to find our desired probability

$ P(g=1|t=1)_{symptom} = \frac{(0.58)*(0.0683)}{(0.011)*(0.9317)+(0.58)*(0.0683)} $


$ P(g=1|t=1)_{symptom} = 0.7945 $

**Here we found that the probability of having COVID if you tested positive and are not experiencing symptoms is roughly 79%**
